```python
import re
import pandas as pd
import unicodedata
```

```python
filename = 'sentimentVN.csv'
df = pd.read_csv(filename)
print(df.shape)
```

```python

filename_list=[]
with open(filename) as file:
  for line in file:
      a = line.lower()
      a = unicodedata.normalize('NFC', a) # chuyển đổi unico tổ hợp -> dựng sẵn
      
      a = a.rstrip('\n')
      a = re.sub(r"[^a-zA-Z0-9đàáảãạâầấẩẫậăằắẳẵặèéẻẽẹêềếểễệđìíỉĩịòóỏõọôồốổỗộơờớởỡợùúủũụưừứửữựỳýỷỹỵ]"," ",a)
      a = re.sub(r'\s+', ' ', a).strip()# xóa khoản trắng dư thừa
      filename_list.append(a )
```

```python
data = df.values
print(data.shape)
lines = data[:,1]
label = data[:,2]
```

```python
print(lines[:5])
print(label[:5])
```

```python
from underthesea import word_tokenize
import nltk 
import numpy as np
```

```python
sentences = []
labels=[]
for i in range(len(lines)):
  try:
    sentences.append(word_tokenize(lines[i]))
    labels.append(word_tokenize(label[i]))
  except:
    print(i)
```

```python
# tách từ trong từng câu 
print(sentences[:5])
print(labels[:15])
print(sentences[:15])
print(len(sentences))
print(len(labels))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
# kiểm tra từ điểm tạo trong tokenizer
#print(tokenizer.word_index)
```

```python
#texts_to_matrix(tokenizer, texts, mode = c("binary", "count", "tfidf", "freq"))
X = tokenizer.texts_to_matrix(sentences,mode='binary')
print(X.shape)
print(X[0])
```

```python
# chia tập train và tập test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, labels, 
                                            test_size=0.2, random_state=0)
```

```python
print(len(y_test))
print(len(X_test))
```

```python
print(y_train)
```

```python
from sklearn.preprocessing import LabelEncoder
lbencoder = LabelEncoder()
y_train2 = lbencoder.fit_transform(y_train)
print(y_train2.shape)
print(y_train2[:14])
```

```python
from tensorflow.keras.utils import to_categorical
y_train2 = to_categorical(y_train2, num_classes=3 )
print(y_train2.shape)
print(y_train2[:3,:])
```

```python
print(X.shape[1])
```

```python
# sử dụng Neural Networks 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(20, input_dim=X.shape[1], activation='sigmoid'))
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(20, activation='sigmoid'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
#train 
model.fit(X_train, y_train2, epochs=20, batch_size=100)
```

```python
y_test1 = lbencoder.fit_transform(y_test)
y_test2 = to_categorical(y_test1, 3)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred[:10])
```

```python
y_pred1 = np.argmax(y_pred,axis=1)
print(y_pred1[:100])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test1,y_pred1))
```
